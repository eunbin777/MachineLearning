```python
cmaps = list(MP.cm.cmap_d.keys())

x = np.random.uniform(-10, 10, size = (100))
y = np.random.uniform(-10, 10, size = (100))
r = np.random.uniform(12, 40, size = (100))
c = np.random.uniform(size = (100))

fig, ax = plt.subplots(1, 1, figsize = (8, 8))

cmap = np.random.choice(cmaps)
ax.set_title(f'Color map : {cmap}')
ax.scatter(x, y, r, c, cmap = cmap)

fig.show()
```

```python
histypes = ['bar', 'barstacked', 'step', 'stepfilled']
fig = plt.figure(figsize = (10, 7))

for n, (col, df) in enumerate(clf_df.iloc[:, :4].iteritems(), 1):
    ax = fig.add_subplot(2, 2, n)

    if n == 1:
        ax.hist(
            x = df,
            bins = np.arange(df.min(), df.max(), .2),
            density = False,
        )
    else:
        for t, sdf in clf_df.groupby('target'):
            ax.hist(
                x = sdf[col],
                bins = np.arange(df.min(), df.max(),  .2),
                density = False,
                label = t,
                histtype = histypes[n-1]
            )
        ax.legend()
        ax.set_title('/'.join([col, histypes[n-1]]))
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15, 5))

tlists = []
for n, (t, sub_df) in enumerate(clf_df.groupby('target')):
    ax.boxplot(
        x = sub_df.iloc[:, :4].dropna().values,
        positions = np.arange(1, 5) - (n * .3),
        widths = .2,
    )
    tlists.append(t)

ax.set_xticklabels(tlists * 4)

tay = ax.twiny()
tay.set_xlim(ax.get_xlim())
tay.set_xticks(np.arange(1, 5) - .3)
tay.set_xticklabels(clf_df.columns[:-1])

fig.show()
```

```python
renamed_reg_df = reg_df.rename(
    columns = {
        i : f'Feature_{i}' if i != 53 else 'Target' for i in reg_df.columns
    })
fig, ax = plt.subplots(1, 2, figsize = (10, 4))

sbn.kdeplot(data = renamed_reg_df, x = 'Target', ax = ax[0])
m, s = renamed_reg_df['Target'].mean(), renamed_reg_df['Target'].std()
ax[0].set_title(f'Mean: {m: .4g} Std. : {s: .4g}')

sbn.kdeplot(x = np.log1p(renamed_reg_df['Target']), ax = ax[1])
m = np.log1p(renamed_reg_df['Target']).mean()
s = np.log1p(renamed_reg_df['Target'].std())
ax[1].set_title(f'Mean: {m: .4g}, Std.: {s: .4g}')
```

```python
logscaled_reg_df = renamed_reg_df.copy()
logscaled_reg_df[['Target', 'Feature_0']] = np.log1p(
    logscaled_reg_df[['Target', 'Feature_0']]
)

sbn.jointplot(
    data = logscaled_reg_df,
    x = 'Feature_0',
    y = 'Target',
    alpha = .3
)
```

```python
def bit2int(x):
    return sum(2 ** n * v for n,v in enumerate(x))

logscaled_reg_df['target0'] = logscaled_reg_df['Target'].apply(lambda x: '>=0' if x != 0 else '==0')
logscaled_reg_df['bit2int'] = logscaled_reg_df.iloc[:, 39:53].apply(lambda x: str(bit2int(x)).rjust(7, '0'), axis = 1)

fig, ax = plt.subplots(figsize = (30, 5))
sbn.boxplot(x = "bit2int", y = "Feature_0", hue = "target0", data = logscaled_reg_df, ax = ax, width = .5)
```
